```python
# Import necessary libraries
import pandas as pd
from IPython.display import display, clear_output
```

```python
# Set paths to cleaned data files
fraud_data_path = "C:/Users/hayyu.ragea/AppData/Local/Programs/Python/Python312/fraud_detection_project/data/processed/fraud_cleaned.csv"
merged_data_path = "C:/Users/hayyu.ragea/AppData/Local/Programs/Python/Python312/fraud_detection_project/data/processed/fraud_with_country.csv"
```

```python
# Load the datasets
print("Loading datasets...")
fraud_df = pd.read_csv(fraud_data_path)
merged_df = pd.read_csv(merged_data_path)
print("Datasets loaded successfully!")

# Display the loaded datasets
print("Displaying the first few rows of the fraud dataset:")
display(fraud_df)  # Show the entire fraud DataFrame

print("\nDisplaying the first few rows of the merged dataset:")
display(merged_df)  # Show the entire merged DataFrame
```

```python
# Convert 'purchase_time' and 'signup_time' to datetime
print("Converting purchase_time and signup_time to datetime...")
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])
print("Conversion completed!")

# Display the updated fraud dataset with converted datetime columns
print("Displaying the updated fraud dataset with datetime columns:")
display(fraud_df)  # Show the updated fraud DataFrame
```

```python
# Feature Engineering
display("Starting feature engineering...")
```

```python
# 1. Transaction Frequency: Number of transactions per user
print("Calculating transaction frequency...")
fraud_df['transaction_count'] = fraud_df.groupby('user_id')['user_id'].transform('count')
print("Transaction frequency calculated.")

# Display the updated fraud dataset with the new 'transaction_count' column
print("Displaying the updated fraud dataset with transaction frequency:")
display(fraud_df)  # Show the updated fraud DataFrame
```

```python
# 2. Transaction Velocity: Average time between transactions for each user
print("Calculating transaction velocity...")

# Sort the DataFrame by 'user_id' and 'purchase_time'
fraud_df = fraud_df.sort_values(by=['user_id', 'purchase_time'])

# Calculate the time difference between consecutive transactions for each user
fraud_df['time_diff'] = fraud_df.groupby('user_id')['purchase_time'].diff().dt.total_seconds()

# Calculate transaction velocity
fraud_df['transaction_velocity'] = fraud_df['time_diff'].fillna(0)

print("Transaction velocity calculated.")

# Display the updated fraud dataset with the new 'transaction_velocity' column
print("Displaying the updated fraud dataset with transaction velocity:")
display(fraud_df)  # Show the updated fraud DataFrame
```

```python
## 3. Time-Based Features
### i. Hour of Day
display("Extracting hour of day...")
fraud_df['hour_of_day'] = fraud_df['purchase_time'].dt.hour

# Display the first few rows of the DataFrame to verify the new feature
display(fraud_df[['purchase_time', 'hour_of_day']].head())
```

```python
### ii. Day of Week (Monday=0, Sunday=6)
display("Extracting day of week...")
fraud_df['day_of_week'] = fraud_df['purchase_time'].dt.dayofweek

# Display the first few rows of the DataFrame to verify the new feature
display(fraud_df[['purchase_time', 'day_of_week']].head())
```

```python
# Save the updated dataset with new features
feature_engineered_output_path = "C:/Users/hayyu.ragea/AppData/Local/Programs/Python/Python312/fraud_detection_project/data/processed/fraud_with_features.csv"
fraud_df.to_csv(feature_engineered_output_path, index=False)

# Display confirmation of the saved dataset
print(f"Feature engineered dataset saved to: {feature_engineered_output_path}")

# Optional: Display the first few rows of the saved DataFrame for verification
print("Displaying the first few rows of the updated fraud dataset:")
display(fraud_df.head())  # Show the first few rows of the updated fraud DataFrame
```
